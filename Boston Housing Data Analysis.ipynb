### T-test for median value of houses bounded and unbounded by the Charles River 

__H0__= There is no significant difference between the means for Median House Value of areas bounded by the Charles River and areas not bounded by the Charles River.

$Alpha$ value=.05

```python
scipy.stats.levene(boston_df[boston_df['CHAS']==1]['MEDV'], boston_df[boston_df['CHAS']==0]['MEDV'], center='mean')
```

_Levene's Test shows the variance is not equal between samples (p-value<alpha), therefore, unequal variances must be accounted for in the t-test._

```python
scipy.stats.ttest_ind(boston_df[boston_df['CHAS']==1]['MEDV'], boston_df[boston_df['CHAS']==0]['MEDV'], equal_var=False)
```

__Conclusion:__ The P-value returned by the t-test is smaller than the alpha value (.05), therefore I __reject the null hypothesis.__ The difference in median value for houses for the two types of locations __is statistically significant.__

### ANOVA test for median value of houses based on proportions of owner-occupied units built before 1940

__H0__= There is no difference in the means of subgroups of median home values when divided by the proportion of old homes.

$Alpha$ value=.05

```python
newhomes=boston_df[boston_df['age_group']=='New Homes']['MEDV']
olderhomes=boston_df[boston_df['age_group']=='Older Homes']['MEDV']
oldesthomes=boston_df[boston_df['age_group']=='Oldest Homes']['MEDV']
```

```python
scipy.stats.levene(newhomes, olderhomes, oldesthomes, center='mean')
```

_Levene's Test shows equality of variance (p-value>alpha), therefore ANOVA test can be performed._

```python
scipy.stats.f_oneway(newhomes, olderhomes, oldesthomes)
```

__Conclusion:__ Since the p-value produced by the ANOVA test is far below the alpha value of .05, we must __reject the null hypothesis.__ When grouped by the proportion of houses built before 1940, the difference in the averages of the median value of houses __is statistic significant.__

### Pearson Correlation Test of Nitric Oxide Concentration and Non-retail Business Acres/Town

__H0__= There is no correlation between the amount of nitric oxides and the number of acres dedicated to non-retail business.

$Alpha$ value=.05

```python
sns.scatterplot(x="NOX", y="INDUS", data=boston_df)
```

_The initial scatterplot shows a potential positive correlation between the two variables, industrial land use and nitric oxide concentrations. A Pearson correlation test must be performed to verify this is statistically significant._

```python
scipy.stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])
```

__Conclusion:__ The p-value returned by the Pearson Correlation test is far lower than the stated alpha value (.05), therefore, __the null hypothesis (H0) must be rejected.__ There is a __statistically significant correlation between the two variables__, in a positive direction (the more nitric oxide present, the more non-retail business land in the town, and vice versa). 

### Regression Analysis for distance from Boston employment centers and the median value of homes

__H0__= There is no effect on the median value of homes from their weighted distance from employment centers.

$Alpha$ value=.05

```python
import statsmodels.api as sm
```

```python
sns.scatterplot(x='DIS', y="MEDV", data=boston_df)
```

```python
x=boston_df['DIS']
y=boston_df['MEDV']
x=sm.add_constant(x)
Model=sm.OLS(y,x).fit()
Predictions=Model.predict(x)
Model.summary()
```

```python
import math
```

```python
r_squared=0.062
Pearson_corr_coef= math.sqrt(r_squared)
Pearson_corr_coef
```

__Conclusion__: The p-value produced by the regression (Prob (F-statistic)=1.12e-08) is less than the alpha value (.05), therefore the __null hypothesis is rejected__. There __is a statistically significant correlation between the 2 variables.__

The impact of an additional weighted distance is approximately: __.249__ (the Pearson correlation coefficient for distance from a Boston employment center).
